### c. Kernel
- We will denote by kernel every function $K\in\mathscr{F}\left(E,\mathbb{R}\right)$ satisfying:
$$\forall p\in\mathbb{N},\forall x_1,\dots,x_p\in E,\forall c_1,\dots,c_p\in\mathbb{R},\sum_{i,j\in\{1,\dots,p\}}c_ic_jK(x_i,x_j)\ge 0$$
In other words, $\forall p\in\mathbb{N},\forall x_1,\dots,x_p\in E$, the matrix $\left(\mathcal{K}_{i,j}\right)_{i,j\in\{1,\dots,p\}}=\left(K(x_i,x_j)\right)_{i,j\in\{1,\dots,p\}}$ is positive semi definite
We may consider kernels as a generalization of inner products

```python
class LinearSupportVectorClassifier:
    def __init__(self,alpha=1):
        self.alpha=alpha
    
    def decision_function(self,X):
        return X@self.w[1:]-self.w[0]
    
    def cost(self,X,Y,w,method=0):
        if method==0:
            return np.sum(np.maximum(0,1-Y*(X@w[1:]-w[0])))+self.alpha*np.linalg.norm(w[1:])
        else: 
            jac=np.r_[0,self.alpha*w[1:X.shape[1]+1]/np.linalg.norm(w[1:X.shape[1]+1]),np.ones(X.shape[0])]
            return self.alpha*np.linalg.norm(w[1:X.shape[1]+1])+np.sum(w[X.shape[1]:]),jac
    def fit(self,X,Y,method=0):
        if method==0:
            R=opt.minimize(lambda w:self.cost(X,Y,w),x0=np.zeros(X.shape[1]+1),method = 'Nelder-Mead')
            self.w=R.x
        else:
            zeta_coeff=np.ones([X.shape[0]]*2)
            w_coeff=np.c_[Y,Y*X[:,0,np.newaxis]]
            A=np.c_[w_coeff,zeta_coeff]
            scipy.optimize.LinearConstraint(A,lb=1,ub=np.inf)
            self.w=opt.minimize(lambda w:self.cost(X,Y,w,method=1),jac = True,x0=np.zeros(X.shape[1]+1+X.shape[0]),method="Newton-CG").x[:X.shape[1]+1]
        return self
    def predict(self,X):
        return np.sign(self.decision_function(X))
    def parameters(self):
        return self.w
    def score(self,X,Y):
        return (self.predict(X)==Y).mean()

class SupportVectorClassifier:
    def __init__(self,alpha=1,K= lambda U,V:U@V.T):
        self.alpha=alpha
        self.K=K
    def decision_function(self,X):
        return np.sign(self.K(X,self.X)@(self.Y*self.w[1:])-self.w[0])
    
    def cost(self,X,Y,w):
        
        return np.sum(np.maximum(0,1-Y*(self.K1@(self.Y*w[1:])-w[0])))+self.alpha*np.linalg.norm(w[1:])
    def fit(self,X,Y):
        self.X=X
        self.K1=self.K(X,X)
        self.Y=Y
        R=opt.minimize(lambda w:self.cost(X,Y,w),x0=np.zeros(X.shape[0]+1),method = "Nelder-Mead")
        self.w=R.x
        return self
    def predict(self,X):
        return np.sign(self.decision_function(X))
    def parameters(self):
        return self.w
    def score(self,X,Y):
        return (self.predict(X)==Y).mean()
```

## Linear Seperation

```python
n1=20
n2=30
X=np.r_[np.random.multivariate_normal([-1,-1],[[3,1],[0,.5]],n1),np.random.multivariate_normal([2,2],[[5,0],[-1,3]],n2)]
Y=np.array([(i<n1)*2-1 for i in range(n1+n2)])
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)
fig,ax = plt.subplots(5,sharex=True,sharey=True,figsize=(16,16))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y,ax=ax[0],palette=["g","r"]);
ax[0].set_title("Lineary separable data")
alpha_range=[0,0.01,1,100]
linear_svc=[LinearSupportVectorClassifier(u) for u in alpha_range]

for i in range(4):
    linear_svc[i].fit(X_train,Y_train,method=1)
    sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=Y_test,ax=ax[i+1],palette=["g","r"]);
    x_limits=np.array([np.min(X[:,0]),np.max(X[:,0])])
    sns.lineplot(x_limits,(linear_svc[i].w[0]-linear_svc[i].w[1]*x_limits)/linear_svc[i].w[2],ax=ax[i+1],color="k")
    sns.lineplot(x_limits,(linear_svc[i].w[0]-linear_svc[i].w[1]*(x_limits-1))/linear_svc[i].w[2],ax=ax[i+1],color="g")
    sns.lineplot(x_limits,(linear_svc[i].w[0]-linear_svc[i].w[1]*(x_limits+1))/linear_svc[i].w[2],ax=ax[i+1],color="r")
    ax[i+1].set_title("$\\alpha={}$\tAccuarcy: {}".format(alpha_range[i],linear_svc[i].score(X_test,Y_test)))
```

```python
n1=200
n2=200
R=np.r_[np.random.normal(1,1,n1),np.random.normal(10,2,n2)]
A=np.random.uniform(0,2*np.pi,n1+n2)
X=np.zeros([n1+n2,2])
X[:,0]=R*np.cos(A)
X[:,1]=R*np.sin(A)

#X=np.c_[X[:,0]*np.cos(X[:,1]),X[:,0]*np.sin(X[:,1])]
Y=np.array([(i<n1)*2-1 for i in range(n1+n2)])
X_train,X_test,Y_train,Y_test=train_test_split(X,Y)

fig,ax = plt.subplots(5,sharex=True,sharey=True,figsize=(8,16))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=Y,ax=ax[0],palette=["g","r"]);
ax[0].set_title("Lineary separable data")
alpha_range=[0,0.01,1,100]
svc=[SupportVectorClassifier(u,K=lambda U,V:np.exp(-U@V.T)) for u in alpha_range]

for i in range(4):
    svc[i].fit(X_train,Y_train)
    sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=Y_test,ax=ax[i+1],palette=["g","r"]);
    x_limits=np.array([np.min(X[:,0]),np.max(X[:,0])])
  
    ax[i+1].set_title("$\\alpha={}$\tAccuarcy: {}".format(alpha_range[i],svc[i].score(X_test,Y_test)))
    ax[i+1].axis('square')
```
